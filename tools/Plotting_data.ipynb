```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the file
data = np.loadtxt('../build/Re_error.line')

# Extract x and y values
x = data[:, 0]
y = data[:, 1]

# Plot the data
plt.plot(x, y, label='Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data Plot')
plt.legend()
plt.grid(True)
plt.show()
```
